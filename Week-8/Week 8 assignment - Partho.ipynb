```python
import nltk
from nltk.corpus import stopwords
import matplotlib.pyplot as plt

%matplotlib inline
```

## 1. Choose a corpus of interest.

The text I will be looking at is the script: **The Legend of Lightning Larry** By Aaron Shepard downloaded from http://www.aaronshep.com/rt/RTE01.html.

The steps followed are:

* Clean data to make it ready for procesing (outside of this program), output file TheLegendofLightningLarryMod.txt
* read the file from disk
* tokenize the text
* normalize the text
    * Make sure everything is in lower case, since we don't want to distinguish between CITIZEN and Citizen.
    * get rid of all punctuation (we do this with w.isalpha())

```python
fl = "TheLegendofLightningLarryMod.txt"
with open(fl, 'r') as fn:
    raw = fn.read()
raw
```

```python
tokens = nltk.word_tokenize(raw)
words = [w.lower() for w in tokens if w.isalpha()]
print "We are having a corpus of {} words in \"The Legend of Lightning Larry\".".format(len(words))
```

```python
words
```

## 2. How many total unique words are in the corpus?

Now let us see how many unique words we have. Let me define unique words as words which are not included in common words such as {a, i, me, we, my, ...} - commonly known as stopwords. We will use nltk feature to remove these words.

```python
uniqueWords = set(words)
uniqueWords = [w for w in uniqueWords if w not in stopwords.words('english')]
print "There are {} unique words in \"The Legend of Lightning Larry\".".format(len(uniqueWords))
```

## 3. Taking the most common words, how many unique words represent half of the total words in the corpus?

Since I have removed all stop words from my unique word list, I should first remove all stop words from the corpus. Then we will see the number of unique words we need to make up half the corpus.

```python
noStopWords = [w for w in words if w not in stopwords.words('english')]
corpusCount = len(noStopWords)
print "We are having a corpus of {} no stop words in \"The Legend of Lightning Larry\".".format(corpusCount)
```

```python
# Frequency Distribution
fdist = nltk.FreqDist([w for w in words if w in uniqueWords])
fdist
```

```python
cumSum = 0
for i, (w, n) in enumerate(fdist.most_common()):
    cumSum += n
    if cumSum > corpusCount / 2:
        numUniqueWords = i
        break

print "{} of the most common words make up half of the corpus".format(numUniqueWords)
```

## 4. Identify the 200 highest frequency words in this corpus.

200 highest frequency words from the no stop word corpus:

```python
fdist.most_common()[:200]
```

200 highest frequency words from the original corpus:

```python
fdistAll = nltk.FreqDist(words)
fdistAll.most_common()[:200]
```

## 5. Create a graph that shows the relative frequency of these 200 words.

```python
maxCount = fdistAll.most_common()[0][1]
relFreq = [1.0 * i[1] / maxCount for i in fdistAll.most_common()]

fig, ax = plt.subplots(figsize=[12, 4])
ax.plot(relFreq[:200])
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel("Word Rank (log scale)")
ax.set_ylabel("Relative Frequency (log scale)")
ax.set_title("Relative Frequency vs Work Rank")
```

## 6. Does the observed relative frequency of these words follow Zipfs law? Explain.

```python
b = 1
x = range(1, 201)
z = [1.0/i**b for i in x]

fig, ax = plt.subplots(figsize=[12, 4])
ax.plot(x, relFreq[:200])
ax.plot(x, z, color='red')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel("Word Rank (log scale)")
ax.set_ylabel("Relative Frequency (log scale)")
ax.set_title("Relative Frequency vs Work Rank")
ax.legend(['actual', 'Zipf'], loc='upper right')
```

The above plot shows that our data set does not exactly follow Zipf's law. Let us see the situation using no stop words corpus.

```python
maxCount = fdist.most_common()[0][1]
relFreq = [1.0 * i[1] / maxCount for i in fdist.most_common()]

fig, ax = plt.subplots(figsize=[12, 4])
ax.plot(relFreq[:200])
ax.plot(x, z, color='red')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel("Word Rank (log scale)")
ax.set_ylabel("Relative Frequency (log scale)")
ax.set_title("Relative Frequency vs Work Rank for no-stop words")
ax.legend(['actual', 'Zipf'], loc='upper right')
```

So our data set does not exactly follow Zipf's law.
